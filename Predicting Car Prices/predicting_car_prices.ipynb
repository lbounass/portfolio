# Introduction to Dataset

```python
import pandas as pd
import numpy as np

cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

cars = pd.read_csv("imports-85.data", names = cols)

cars
```

```python
cars.info()
```

```python
numeric_values = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[numeric_values]
numeric_cars.head(10)
```

# Data Cleaning

```python
numeric_cars = numeric_cars.replace("?", np.nan)
numeric_cars = numeric_cars.astype('float')
numeric_cars.info()
```

```python
numeric_cars.isnull().sum()
```

```python
#First, let's remove the rows where price is nan:
numeric_cars = numeric_cars [~ np.isnan(numeric_cars['price'])]
numeric_cars.head(10)
```

```python
#As for the rest, we replace nan values with the mean:
numeric_cars = numeric_cars.fillna(numeric_cars.mean())
numeric_cars.isnull().sum()
```

```python
#Now we normalize:
target_col = numeric_cars ['price']
numeric_cars = (numeric_cars.max() - numeric_cars)/(numeric_cars.max())
numeric_cars['price'] = target_col
```

```python
numeric_cars
```

# Univariate Model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

#####For k = 5
features = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']
target = ['price']
univariate_rms = {}

def knn_train_test(training_col, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    
    border = np.ceil(rand_df.shape[0]*0.5).astype('int')
    training_set = rand_df[0: border ]
    test_set = rand_df [border : ]
    knn = KNeighborsRegressor()
    knn.fit (training_set [training_col] , training_set [target_col])
    predictions = knn.predict (test_set[training_col])
    rms = np.sqrt(mean_squared_error (test_set['price'], predictions))
    return rms
    
for f in features:
    univariate_rms [f] = knn_train_test([f], target, numeric_cars)
    
print (univariate_rms)

min_rms = univariate_rms ["horsepower"]
min_feature = "horsepower"

for f in features:
    if univariate_rms[f] < min_rms:
        min_rms = univariate_rms[f]
        min_feature = f
print ("Best-performing column:", min_feature, "with rms=", min_rms)
        

```

```python
###For different parameters of k

features = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']
target = ['price']
univariate_rms = {}
k_values = [1, 3, 5, 7, 9]

def knn_train_test(training_col, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    rms = {}
    border = np.ceil(rand_df.shape[0]*0.5).astype('int')
    training_set = rand_df[0: border ]
    test_set = rand_df [border : ]
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit (training_set [training_col] , training_set [target_col])
        predictions = knn.predict (test_set[training_col])
        rms [k]= np.sqrt(mean_squared_error (test_set['price'], predictions))
    return rms
    
for f in features:
    univariate_rms [f] = knn_train_test([f], target, numeric_cars)
    
univariate_rms
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for f in features:
    temp_rms = []
    size = len(univariate_rms[f])
    for k in k_values:
        temp_rms.append(univariate_rms[f][k])
    print ("For ", f, ":")
    plt.scatter(k_values, temp_rms)
    plt.show()
```

# Multivariate Model

```python
#To find the best features, we calculate the average rms for each feature:
rms_avg = dict()
for f in features:
    temp_rms = []
    size = len(univariate_rms[f])
    for k in k_values:
        temp_rms.append(univariate_rms[f][k])
    rms_avg [f] = np.mean(temp_rms)
series_rms_avg = pd.Series(rms_avg) 
sorted_rms_avg = series_rms_avg.sort_values()
sorted_rms_avg
```

# Hyperparameter Tuning

```python
#The top 3 models are 2, 3 and 4
multivariate_rms = {}
k_values = [i for i in range (1, 25)]
best_three = [2, 3, 4]

def knn_train_test(training_col, target_col, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    rms = {}
    border = np.ceil(rand_df.shape[0]*0.5).astype('int')
    training_set = rand_df[0: border ]
    test_set = rand_df [border : ]
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit (training_set [training_col] , training_set [target_col])
        predictions = knn.predict (test_set[training_col])
        rms [k]= np.sqrt(mean_squared_error (test_set['price'], predictions))
    return rms
    
for f in best_three:
    multivariate_rms [f] = knn_train_test(sorted_rms_avg.index[0:f], target, numeric_cars)

    
multivariate_rms
```

```python
for f in best_three:
    temp_rms = []
    size = len(multivariate_rms[f])
    for k in k_values:
        temp_rms.append(multivariate_rms[f][k])
    print ("\nFor ", f, ":")
    plt.scatter(k_values, temp_rms)
    plt.show()
    
    min_temp_rms = temp_rms [0]
    min_k = 1
    
    for i in range (0, len(temp_rms)):
        if (temp_rms[i] < min_temp_rms):
            min_temp_rms = temp_rms [i]
            min_k = i + 1
    print ("k =", min_k, "is optimal for model #", f, "with rmse=", min_temp_rms)
```
