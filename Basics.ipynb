# Converting Age To Numeric

```python
def age_conversion(string_age):
    if pd.isnull(string_age):
        return None
    else:
        new_string = (string_age.split(" "))[0]
        if (new_string[-len(new_string)] == "+"):
            return int(new_string.split("+")[0])
        else:
            return  int(new_string[0:2])
data["int_age"] = (data["Age"].apply(age_conversion))   
print(data["int_age"])
print(data["int_age"].describe())
print(data["Age"].value_counts())
```

# Findings and Analysis

Even though we can see that the age groups are more or less evenly distributed, the fact that we have assigned every age corresponding to a certain range to the lower bound has caused the mean to be lower than it shuold be.

Also, we should keep in mind that some values are NaN, which makes some of the findings somewhat less reliable. The question remains on what we should do with these missing values, drop these rows completely or assign an "average" value?

# Converting Income To Numeric

```python
def income_earned(income):
    if pd.isnull(income):
        return None
    else:
        first_item= income.split(" ")
        if (first_item[0]=="Prefer"):
            return None
        else :
            without_dollar = first_item[0].replace("$", "")
            without_dollar_and_comma = without_dollar.replace(",", "")
            return int(without_dollar_and_comma)
data["int_income"] = data["How much total combined money did all members of your HOUSEHOLD earn last year?"].apply(income_earned)
#print(data["int_income"])  
print(data["int_income"].describe())
data["How much total combined money did all members of your HOUSEHOLD earn last year?"].value_counts()

```

# Findings and Analysis

Similarly to the age, the mean is skewed downwards due to our assigning all income to the lowest bound value of the range. Other than that, it is a true depiction of the incomes of survey participants. 

It is worth to note, nevertheless, that the average is relatively large when the highest income group ($25,000  to $49,999), considerably surpasses the other groups, but that can be explained by the high standard deviation.

# Correlating Travel Distance And Income

```python
data_income_lt_150000 = data [ data ["int_income"] <150000 ]
print ("For people earning less than $150,000:")
print("\n", data_income_lt_150000 [ "How far will you travel for Thanksgiving?" ].value_counts())
print("\n", "In Percentage (%):")
print("\n", data_income_lt_150000 [ "How far will you travel for Thanksgiving?" ].value_counts()/len(data_income_lt_150000)*100)

print ("\nFor people earning more than $150,000:")
data_income_gt_150000 = data [ data ["int_income"] >150000 ]
print("\n", data_income_gt_150000 [ "How far will you travel for Thanksgiving?" ].value_counts())
print("\n", "In Percentage (%):")
print("\n", data_income_gt_150000 [ "How far will you travel for Thanksgiving?" ].value_counts()/len(data_income_gt_150000)*100)



print("\n", "By income groups: \n", data["int_income"].value_counts())
```

# Findings and Analysis

We see that for people earning less than 150,000 dollars, around 41 percent are not willing to travel at all, while 48 percent of people earning more than 150,000 dollars prefer not to move. Thus, it is reasonable to believe that people earning more than 150,000 dollars are more prone to host Thanksgiving at their residence than those earning less. 

Additionaly, those earning less are more willing to travel for a few hours than those earning more (if Thanksgiving is out of town but not too far). 

However, if Thansgiving is far away, those earning less are less prone to travel than those earning more, likely due to moving and traveling costs.

The fact that people earning less than 150,000 dollars are not very prone to host Thanksgiving but are willing to travel only if Thanksgiving isn't located too far seems to suggest that these people tend to be young people traveling to their hometown to celebrate Thanksgiving with their families. Conversely,the people who earn more and yet are less likely to travel small distances seem to fit the profile of parents who host Thanksgiving in their house.

In order to confirm that the Thanksgiving hosts are more likely to be parents and that those traveling are more likely to be young people traveling to their hometown, it'd be interesting to analyze the relationship between the "Age" column and the "int_income" column.

# Linking Friendship And Age

```python
data.pivot_table(values = "int_age",index = "Have you ever tried to meet up with hometown friends on Thanksgiving night?", columns = 'Have you ever attended a "Friendsgiving?"')
```

```python
data.pivot_table(values = "int_income",index = "Have you ever tried to meet up with hometown friends on Thanksgiving night?", columns = 'Have you ever attended a "Friendsgiving?"')
```

# Findings and Analysis

In the first pivot table, we notice that people who don't attend Friendsgiving are on average 41-42 years old, an average much larger that that of people who do attend Friendsgiving. 

Looking at the index "Have you ever tried to meet up with hometown friends on Thanksgiving night?", we can already assume that the people who responded "Yes" to that question either live in their hometown or have travelled back home. We observe that the people who celebrate Thanksgiving and meet up hometown friends are much younger than those who meet up with homewtown friends but don't celebrate Friendsgiving. This suggests that younger people are more prone to celebrate Friendsgiving than older people (such as parents), regardless of the actual location of Friendsgiving. This thus explains why the people who don't celebrate Friendsgiving, whether they meet up with hometown friends or not, are around the same age group. Additionally, the analysis also suggests that any activity related to meeting friends (whether it is Friendsgiving or meeting up hometown friends), is usually performed by younger people, while older people's Thanksgiving revolves more around family members.

We can also observe a correlation between the age of the participants and their income. The younger they are, the less income they earn. Thus, the older age group, conveniently the higher income group, neither celebrate Friendsgiving nor meet up with friends during Thanksgiving, reinforcing the hypothesis that those people already live in their hometown and host Thanksgiving themselves. Conversely, the younger people, and those who earn less, are much more likely to celebrate Friendsgiving and travel back to their hometown to celebrate Thanksgiving with their families.
